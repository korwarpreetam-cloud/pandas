# group by aggregation 

```python
import numpy as np 
import pandas as pd
```

```python
data = {
    'Category': ['A','B','A','B','A','B','A','B'],
    'Store':['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Quantity':[10,15,12,18,8,20,15,25],
    'Date':pd.date_range('2023-01-01', periods=8)



}
df = pd.DataFrame(data)
df
```

```python
# Group by store and calculate the sum pf the sales
cat = df.groupby('Category')['Quantity'].sum()
print(cat)


```

```python
# group by multiple columns
cat = df.groupby(['Category','Store'])['Quantity'].sum()
print(cat)
```

```python
df['Quantity'].mean()
#median mean  min max count std 
```

```python
df['Quantity'].agg(['mean','min','max'])
```

## pivot tables


```python


data = [
    ['2023-01-01', 'A', 'East', 825, 26, 'John', 'January', 'Q1'],
    ['2023-01-02', 'B', 'West', 611, 51, 'Mary', 'January', 'Q1'],
    ['2023-01-03', 'C', 'North', 768, 11, 'Bob', 'January', 'Q1'],
    ['2023-01-04', 'D', 'South', 665, 71, 'Alice', 'January', 'Q1'],
    ['2023-01-05', 'A', 'East', 157, 20, 'John', 'January', 'Q1'],
    ['2023-01-06', 'B', 'West', 436, 43, 'Mary', 'January', 'Q1'],
    ['2023-01-07', 'C', 'North', 227, 39, 'Bob', 'January', 'Q1'],
    ['2023-01-08', 'D', 'South', 783, 35, 'Alice', 'January', 'Q1'],
    ['2023-01-09', 'A', 'East', 352, 54, 'John', 'January', 'Q1'],
    ['2023-01-10', 'B', 'West', 740, 31, 'Mary', 'January', 'Q1']
]

columns = ['Date', 'Product', 'Region', 'Sales', 'Units', 'Rep', 'Month', 'Quarter']

df = pd.DataFrame(data, columns=columns)

print(df)
```

```python
pd.pivot_table(df,values="Sales",index='Region',columns='Product',aggfunc='median')
```

```python
pivot2=pd.pivot_table(df, values=['Sales','Units'],index='Region',columns='Product')
pivot2
```
