### find missing data

### merging joining and contcanation

### merging

```python
employees = pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'name':['john','anna','peter','linda','bob'],
    'deparment': ('Hr','it','finance','it','hr')
})

salaries = pd.DataFrame({
    'employee_id':[1,2,3,5,6],
    'salary':[60000,80000,90000,30000,10000],
    'bonus':[5000,1000,3000,4000,7000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how='inner')
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer')
```

```python
pd.merge(employees,salaries,on='employee_id',how='left')
```

```python
pd.merge(employees,salaries,on='employee_id',how='right')
```

### concatation

```python
pf1 = pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2'],
    'C':['C0','C1','C3']
})
pf2 = pd.DataFrame({
    'A':['A3','A2','A3'],
    'B':['B3','B2','B0'],
    'C':['C3','C2','C1']

})
pf1
pf2
```

```python
pf1
```

```python
pf2
```

```python
pd.concat([pf1,pf2])
```

```python
pd.concat([pf1,pf2],axis=1)
```

### joinging

```python
fg1=pd.DataFrame({
    'name':['Alice','Bob','charlie'],

}, index=[1,2,3])

fg2=pd.DataFrame({
    'score':[85,90,75]
}, index=[2,3,4])

fg1
```

```python
fg1
```

```python
fg2
```

```python
fg1.join(fg2)
```

```python
fg1.join(fg2,how='outer')
```
